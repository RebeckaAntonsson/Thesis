Create a fasta file from the biophi dataset with ADA and sequence data

```python
# load libaries
import pandas as pd
import numpy as np
```

```python
# load data
ada_biophi_dataset = pd.read_csv("ADA_Clinical_Ab_2021_biophiDataset.csv")
```

```python
# first remove white spaces just in case
ada_biophi_dataset['Antibody'] = ada_biophi_dataset['Antibody'].replace(r"\s+", "", regex=True).str.lower()
ada_biophi_dataset['VH'] = ada_biophi_dataset['VH'].replace(r"\s+", "", regex=True)
ada_biophi_dataset['VL'] = ada_biophi_dataset['VL'].replace(r"\s+", "", regex=True)

ada_biophi_dataset.head()
# create fasta file will all antibodies
lines_normal = []

for _, row in ada_biophi_dataset.iterrows():
    lines_normal.append(f">{row['Antibody']}")
    lines_normal.append(str(row['VH'])+str(row['VL']))

fasta_normal = "\n".join(lines_normal)

with open("fasta_files/ada_biophi_ABdataSet.fasta", "w") as f:
    f.write(fasta_normal)
```

# waltz
waltz can take 10,000 characters at most. This fasta file is ~52,000. So below i divide it into 6 fasta files instead.

```python

# split dataframe into 6 roughly equal parts
chunks = np.array_split(ada_biophi_dataset, 6)

for i, chunk in enumerate(chunks, start=1):
    lines_waltz = []
    
    for _, row in chunk.iterrows():
        lines_waltz.append(f">{row['Antibody']}")
        lines_waltz.append(str(row['VH']) + str(row['VL']))
    
    fasta_waltz = "\n".join(lines_waltz)
    
    with open(f"fasta_files/ada_biophi_ABdataSet_waltz_part{i}.fasta", "w") as f:
        f.write(fasta_waltz)
```

# Biophi
Biophi required a fasta file where the two chains are devided, with a _VH or _VL (heavy chain and light chain) prefix. Maximum number of input sequences are 100, therefore the data set is split into 3. Below I create a fasta file for that.\
Clarification. THe data set is called biophi because it comes from the creation of this tool. I also test the biophi tool here.

```python
# split dataframe into 6 roughly equal parts
chunks = np.array_split(ada_biophi_dataset, 3)

# create fasta file
for i, chunk in enumerate(chunks, start=1):

    lines_biophi = []

    for _, row in chunk.iterrows():
        lines_biophi.append(f">{row['Antibody']}_VH")
        lines_biophi.append(str(row['VH']))
        lines_biophi.append(f">{row['Antibody']}_VL")
        lines_biophi.append(str(row['VL']))


    fasta_biophi = "\n".join(lines_biophi)

    with open(f"fasta_files/ada_biophi_ABdataSet_VH_VL_part{i}.fasta", "w") as f:
        f.write(fasta_biophi)
    
```
