```python
# load libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
timothy_data = pd.read_excel("Timothy_data_set.xlsx")
```

# FASTA with selected minibinders

```python
# filter data

# Get only 20 minibinders for each target
# get 10 binders and 10 non-binders

# taget groupts with less than 10
# split into large and small groups



# filter out 10 of each target and binder
timothy_filtered = (
    timothy_data
    .groupby(["target_id", "binder"], group_keys=False)
    .apply(lambda x: x.sample(n=min(len(x), 10), random_state=42))
)

# keep only columns of interest
# binder_id, target_id, A_seq
timothy_filtered = timothy_filtered[['binder_id', 'target_id', 'A_seq']]
timothy_filtered.head()
# make it into a fasta file

# first remove white spaces just in case
timothy_filtered['binder_id'] = timothy_filtered['binder_id'].replace(r"\s+", "")
timothy_filtered['A_seq'] = timothy_filtered['A_seq'].replace(r"\s+", "")

# create fasta file
lines = []

for _, row in timothy_filtered.iterrows():
    lines.append(f">{row['binder_id']}")
    lines.append(str(row["A_seq"]))

fasta = "\n".join(lines)

with open("filterd_timothy.fasta", "w") as f:
    f.write(fasta)
```
